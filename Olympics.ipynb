### STATISTICAL TESTS

```python
sports['winter_participated'] = sports['winter_participations'].apply(lambda x: 'Yes' if x > 0 else 'No')
print(sports[['countries', 'winter_participations', 'winter_participated']].head())
```

### CHI SQUARE TEST 

```python
import scipy.stats as stats
contingency_table = pd.crosstab(sports['winter_participated'], sports['region'])

chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)


print("Chi-Square Statistic:", chi2_stat)
print("p-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:\n", expected)

alpha = 0.05
print("******************************************")
if p_value < alpha:
    print("REJECT THE NULL HYPOTHESIS")
    print("There is a significant relationship between winter participation and region.")
else:
    print("FAIL TO REJECT THE NULL HYPOTHESIS")
    print("There is no significant relationship between winter participation and region.")
```

### INDEPENDENT TWO SAMPLE  t -TEST

```python
from scipy.stats import ttest_ind

data = sports[['summer_total', 'winter_total']].dropna()

summer_medals = data['summer_total']
winter_medals = data['winter_total']

t_stat, p_value = ttest_ind(summer_medals, winter_medals, equal_var=False)  

print(f'Two-Sample t-Test: t-statistic = {t_stat}, p-value = {p_value}')
alpha = 0.05
print("******************************************")
if p_value < alpha:
    print("REJECT THE NULL HYPOTHESIS")
    print("There is a statistically significant difference.")
else:
    print("FAIL TO REJECT THE NULL HYPOTHESIS")
    print("There is no statistically significant difference.")
```
